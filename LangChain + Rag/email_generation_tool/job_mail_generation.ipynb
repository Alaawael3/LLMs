# Application

```python
def process_job(url):
    # Load and scrape webpage
    loader = WebBaseLoader(url)
    data = loader.load()
    page = data[0].page_content

    # Extract job posting
    chain_extract = prompt_extract | llm
    job_json = chain_extract.invoke({"page_data": page})
    parsed_json = json_parser.parse(job_json.content)
    job_desc = parsed_json['skills']

    # Query Chroma collection
    results = collection.query(query_texts=job_desc, n_results=2)

    # Generate cold email
    chain = prompt_email | llm
    response = chain.invoke({
        "job_description": job_desc,
        "link_list": results.get('metadatas')
    })

    return response.content

# Gradio Interface
interface = gr.Interface(
    fn=process_job,
    inputs=[
        gr.Textbox(label="Job URL"),
        # gr.Textbox(label="Skill Query (e.g., Python, React)")
    ],
    outputs=[
        # gr.JSON(label="Extracted Job JSON"),
        gr.Textbox(label="Generated Cold Email")
    ],
    title="Email Generator"
)

interface.launch()
```
